```python
# tạo biến mới, đổi 1->0 và 0 -> 1
# cách 1
gcd_data.loc[:,'Creditability2'] = 1 - gcd_data.loc[:,'Creditability']
print(gcd_data.loc[:,['Creditability','Creditability2']])
```

```python
# tạo biến mới, đổi 1->0 và 0 -> 1
# cách 1
gcd_data.loc[:,'Creditability2'] = 0
gcd_data.loc[gcd_data['Creditability']==0,'Creditability2'] = 1
print(gcd_data.loc[:,['Creditability','Creditability2']])
```

```python
# Xem cột purpose
gcd_data['Purpose'][:10]
```

```python
# tạo bảng mapping cho cột purpose
purpose_mapping = {0 : 'car (new)', 
1 : 'car (used)',
2 : 'furniture/equipment', 
3 : 'radio/television', 
4 : 'domestic appliances', 
5 : 'repairs',
6 : 'education', 
7 : '(vacation - does not exist?)',
8 : 'retraining', 
9 : 'business', 
10 : 'others'}
# tạo cột mới, với các giá trị được thay theo quy tắc mapping trên
gcd_data.loc[:,'Purpose (Text)'] = gcd_data['Purpose'].map(purpose_mapping)
```

```python
reversed_mapping = {v:k for k, v in purpose_mapping.items() }
gcd_data.loc[:,'Purpose (Text)'].map(reversed_mapping)[:10]
```

```python
gcd_data.groupby('Creditability2').agg({'Account Balance':['mean','std','max'],
                                       'Duration of Credit (month)':['mean','std','min'],}).T
```

```python
def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_

gcd_data.groupby('Creditability2').agg({'Account Balance':['mean','std','max', percentile(95)],
                                       'Duration of Credit (month)':['mean','std','min',percentile(5)],}).T
```

```python
gcd_data.groupby(gcd_data['Purpose'].map(purpose_mapping)).agg({'Account Balance':['mean','std','max', percentile(95)],
                                       'Duration of Credit (month)':['mean','std','min',percentile(5)],}).T
```
